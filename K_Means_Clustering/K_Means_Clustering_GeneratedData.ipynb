```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

**Create some data**

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data[0].shape
```

**plot a scatter between the 2 features**

```python
plt.scatter(data[0][:,0], data[0][:,1])
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1])
```

```python
data[1]
```

**K Means Clustering **

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

**For real data we are done at this stage. we wont have data to compare as it is unsupervised data**

```python
fig, (ax1, ax2) = plt.subplots(1,2, sharey=True,figsize=(10,6))

ax1.scatter(data[0][:,0], data[0][:,1], c=data[1])
ax1.set_title("ACTUAL")

ax2.scatter(data[0][:,0], data[0][:,1], c= kmeans.labels_)
ax2.set_title("K MEANS")
```

**number of clusters = 2**

```python
kmeans = KMeans(n_clusters = 2)
kmeans.fit(data[0])

fig, (ax1, ax2) = plt.subplots(1,2, sharey=True,figsize=(10,6))

ax1.scatter(data[0][:,0], data[0][:,1], c=data[1])
ax1.set_title("ACTUAL")

ax2.scatter(data[0][:,0], data[0][:,1], c= kmeans.labels_)
ax2.set_title("K MEANS")
```
