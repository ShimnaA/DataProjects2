```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'], columns=cancer['feature_names'])
```

```python
df.head()
```

```python
type(cancer)
```

Visualization

WE need to Scale data - Each data to single unit variance

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
```

```python
scalar.fit(df)
```

```python
df_scaled = scalar.transform(df)
```

```python
df_scaled.shape
```

PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(df_scaled)
```

```python
x_pca = pca.transform(df_scaled)
```

```python
x_pca.shape
```

We have reduced the 30 features to 2 features

```python
plt.figure(figsize=(10,6))
plt.scatter(x = x_pca[:, 0], y = x_pca[:, 1], c=cancer['target'])
plt.xlabel = 'First principal component'
plt.ylabel='Second PCA Component'
```

```python
type(x_pca)
```

```python
x_pca
```

```python
pca.components_
```

```python
pca_component_df = pd.DataFrame(pca.components_, columns=cancer['feature_names'])
```

```python
pca_component_df
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(pca_component_df, cmap='plasma')
```

Use PCA to reduce dimension to 3

```python
scalar.fit(df)
```

```python
scaled_data = scalar.transform(df)
```

```python
pca = PCA(n_components=3)
```

```python
pca.fit(scaled_data)
```

```python
x1_pca = pca.transform(scaled_data)
```

```python
x1_pca.shape
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x = x1_pca[:, 0], y = x1_pca[:, 1], c=x1_pca[:, 2])
plt.xlabel = 'First principal component'
plt.ylabel='Second PCA Component'
```
